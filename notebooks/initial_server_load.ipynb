```python
#dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Connect to sql server INSERT DB NAME
rds_connection_string = "postgres:password@35.193.2.240:5432/"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
#Read in initial.csv
df = pd.read_csv(r'../Resources/initial.csv')
df.head()
```

```python
#create df to load into master table
master_df = df[['job_title', 'company', 'salary_range', 'abstract', 'post_url', 'date_added', 'city', 'state']]
master_df.head(10)
```

```python
#rename columns to match table
master_df = master_df.rename(columns={'salary_range': 'salary', 'abstract': 'job_description', 'post_url': 'job_post_url', "date_added": 'post_date', 'state': 'state_abbr'})
master_df.head()
```

```python
#Load into master table
master_df.to_sql(name='master', con=engine, if_exists='append', index=False)
```

```python
#create city table
city_df = pd.DataFrame(df['city'].unique())
city_df.rename(columns = {0: 'city_name'}, inplace=True)
city_df.head()
```

```python
city_df.to_sql(name='city', con=engine, if_exists='append', index=False)
```

```python
conn = engine.connect()
```

```python
city_id_df = pd.read_sql_query('select * from city', con=conn).head()
conn.close()
```

```python
#create state table
state_df = pd.DataFrame(df['state'].unique())
state_df.rename(columns = {0: 'state_abbr'}, inplace=True)
state_df.head()
```

```python
#load into state table
state_df.to_sql(name='state', con=engine, if_exists='append', index=False)
```

```python
state_id_table = pd.read_sql_query('select * from state', con=conn).head()
conn.close()
```
